```python
class Graph:
    def __init__(self,adjacent,heuristics):
        self.adjacent = adjacent
        self.heuristics = heuristics
    
    def get_heighbors(self,node):
        return self.adjacent.get(node,[])

    def h(self,node):
        return self.heuristics.get(node,float('inf'))
```

```python
import heapq
import itertools

class Astart:
    def __init__(self, graph):
        self.graph = graph
        self.counter = itertools.count()

    def search(self, start, goal):

        frontier = []
        heapq.heappush(
            frontier,
            (self.graph.h(start), 0, next(self.counter), start)
        )

        g_cost = {start: 0}
        parents = {start: None}
        visited = set()
        visited_ = []

        print(f"** Search From {start} to {goal} **")
        print(visited)
        while frontier:
            print(frontier)
            f_current, g_current,_, current_node = heapq.heappop(frontier)
            if current_node == goal:
                return self._reconstruc(parents, current_node), g_current

            if current_node in visited:
                continue

            visited.add(current_node)
            visited_.append(current_node)
            print(visited_)
            for neighbor, weight in self.graph.get_heighbors(current_node):
                if neighbor in visited:
                    continue
                
                tentative_g = g_current + weight

                if tentative_g < g_cost.get(neighbor, float('inf')):
                    g_cost[neighbor] = tentative_g
                    parents[neighbor] = current_node
                    f_cost = tentative_g + self.graph.h(neighbor)
                    
                    heapq.heappush(
                        frontier,
                        (f_cost, tentative_g, next(self.counter), neighbor)
                    )

        return None, float('inf')

    def _reconstruc(self, parents, current_node):
        path = []
        while current_node is not None:
            path.append(current_node)
            current_node = parents[current_node]
        return path[::-1]
```

```python
Graph_data = {
    'S': [('A',2),('D',5)],
    'A':[('B',3),('C',4)],
    'D':[('B',3),('E',4)],
    'B':[('C',3),('E',5)],
    'E':[('G',6)],
    'C':[('G',2)],
}

h_data = {
    'S':7,
    'A':9,
    'B':4,
    'C':2,
    'D':5,
    'G':0,
    'E':3,
}

graph = Graph(Graph_data,h_data)
result = Astart(graph)
path , total_cost = result.search('S','G')

print("-"*30)
if path:
    print(f"Shortest path: {' -> '.join(path)}")
    print(f"Total Cost: {total_cost}")
else:
    print("Not Found")
```
