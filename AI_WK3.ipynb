```python
graph = {
    'A': ['B', 'C'],
    'B': ['D', 'E'],
    'C': ['F', 'G'],
    'D': ['H', 'I'],
    'E': ['J'],
    'F': ['K'],
    'G': [],
    'H': ['C'],
    'I': [ ],
    'J': [ 'L'],
    'K': [ ]
}
```

```python

path_current = set()
def recursive_dls(graph, current_node, goal_state, limit, current_depth, visited_nodes, parents):
    
    if current_node == goal_state:
        return current_node
    if current_depth == limit:
        return "Cutoff"
    cutoff_occurred = False
    
    visited_nodes.add(current_node)
    path_current.add(current_node)
    for neighbor in graph.get(current_node, []):
        
        if neighbor not in visited_nodes:
            print(f"Visiting Node: {neighbor} at depth {current_depth + 1}")
            result = recursive_dls(graph, neighbor, goal_state, limit, \
                current_depth + 1, visited_nodes, parents)
            if result == "Cutoff":
                cutoff_occurred = True
                
            elif result is not None:
                print(path_current)
                return result
    
    visited_nodes.remove(current_node)
    if cutoff_occurred:
        print(path_current)
        return "Cutoff"
  
    return None

    
```

```python
def reconstruct_path(parents, current_node):
    path = []
    while current_node is not None:
        path.append(current_node)
        current_node = parents.get(current_node)
    return path[::-1] 
```

```python
def ids_search(graph,start_node,goal_state,max_depth):
    print(f"IDS (Start: {start_node}, Goal: {goal_state}, Max Depth: {max_depth})")
    for depth in range(max_depth + 1):
        print(f"Current search levet at: {depth}")
    
        visited_nodes = set()
        parents = {start_node: None}
        result = recursive_dls(graph, start_node, goal_state, depth, 0, visited_nodes, parents)
        if result is not None and result != "Cutoff":
            print(f"Gold state found at level {depth}")
            return reconstruct_path(parents, result)
    print("Not found gold state until max depth")
    return None
```

```python
solution_path = ids_search(graph, 'A', 'J', max_depth=5)
if solution_path:
    print(f"Search path to goal state: {' -> '.join(solution_path)}")
else:
    print("Not found goal state")
```

```python
import heapq

def greedy_best_first_search_trace(graph, start, goal, heuristic):
    frontier = []
    heapq.heappush(frontier, (heuristic[start], start))

    visited = []
    visited_set = set()

    step = 0

    def print_frontier(frontier):
        return "[" + ", ".join(f"{n}{h}" for h, n in sorted(frontier)) + "]"

    while frontier:
        print(f"• Frontier:{print_frontier(frontier)};")
        print(f"Visited={visited}")

        h, current = heapq.heappop(frontier)

        if current == goal:
            print(f"• Success, {current}= goal!")
            return

        visited.append(f"{current}{h}")
        visited_set.add(current)

        for neighbor in graph.get(current, []):
            if neighbor not in visited_set:
                heapq.heappush(
                    frontier,
                    (heuristic[neighbor], neighbor)
                )

        step += 1
```

```python
graph = {
    'S': ['A', 'B', 'C'],
    'A': ['D', 'E'],
    'B': ['F', 'G'],
    'C': ['H'],
    'H': ['I', 'J'],
    'I': ['K', 'L', 'M'],
    'G': []
}

heuristic = {
    'S': 0,
    'A': 3,
    'B': 6,
    'C': 5,
    'D': 9,
    'E': 8,
    'F': 12,
    'G': 14,
    'H': 7,
    'I': 5,
    'J': 6,
    'K': 1,
    'L': 10,
    'M': 2
}

greedy_best_first_search_trace(graph, 'S', 'I', heuristic)
```
